```python
import requests
from bs4 import BeautifulSoup
from lxml import etree

url =  "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"

response = requests.get(url)
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
```

```python
dom = etree.HTML(str(soup))
rows = dom.xpath('//div[@class="col-md-4 col-xl-4 col-lg-4"]')
title = dom.xpath('//a[@class="title"]/@title')
price = dom.xpath('//h4[@class="float-end price card-title pull-right"]/text()')
des = dom.xpath('//p[@class="description card-text"]/text()')
```

```python
for i in range(len(rows)):
    print(f"name of prod: {title[i]}\n\
price: {price[i]}\n\
description: {des[i]}")
    print("\n\n")
```
